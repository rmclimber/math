#### Exercise 4.1
Compute the determinant using the Laplace expansion (using the first row) and the Sarrus rule for:
\\[
A = \begin{bmatrix}
1 & 3 & 5 \\
2 & 4 & 6 \\
0 & 2 & 4
\end{bmatrix}
\\]

```python
def sarrus(A: np.ndarray) -> np.float32:
    """
    Formula on MML text p. 100, equation 4.7
    """
    m, n = A.shape
    if m != n or m !=3:
        raise ValueError("Not a 3x3")
    addition = A[0, 0] * A[1, 1] * A[2, 2] + A[1, 0] * A[2, 1] * A[0, 2] + A[2, 0] * A[0, 1] * A[1, 2]
    subtraction = -A[2, 0] * A[1, 1] * A[0, 2] - A[0, 0] * A[2, 1] * A[1, 2] - A[1, 0] * A[0, 1] * A[2, 2]
    return addition + subtraction
```

```python
def det_2x2(A: np.ndarray) -> np.float32:
    return A[0, 0] * A[1, 1] - A[0, 1] * A[1, 0]
```

```python
def laplace_det(A: np.ndarray) -> np.float32:
    """
    This allows for generalized computation of the determinant of
    nxn matrices using the Laplace expansion.
    """
    m, n = A.shape
    if m != n:
        raise ValueError("Not a square matrix: determinant undefined")
    # base cases
    if m == 1:
        return A[0, 0]
    elif m == 2:
        return det_2x2(A)
    det = 0
    for i in range(m):
        j = i
        sign = (-1)**j
        row_deleted = np.delete(A, 0, axis=0)
        minor = np.delete(row_deleted, j, axis=1)
        subproblem = laplace_det(minor)
        det += sign * A[0, j] * subproblem
    return det
```

```python
A = np.array(
    [
        [1, 3, 5],
        [2, 4, 6],
        [0, 2, 4]
    ]
)
```

```python
true_det = np.linalg.det(A)
print(f"numpy determinant: {true_det}")
sar_det = sarrus(A)
print(f"Sarrus determinant: {sar_det}")
det_4_1 = laplace_det(A)
print(f"LaPlace determinant: {det_4_1}")
```

#### Exercise 4.3

a. Compute the eigenspace of: \\(A = \begin{bmatrix} 1 & 0 \\ 1 & 1\end{bmatrix}\\). 

Approach: we know that the eigenspace \\(E_\lambda\\) is the solution space for \\((A - \lambda I)x = 0\\), where \\(x\\) is an eigenvector associated with eigenvalue \\(\lambda\\). In other words, we solve for a non-zero \\(x\in\mathbb{R}^2\\) such that \\(A - \lambda I\\) will be mapped to \\(0\\), i.e. \\(x\in\ker(A - \lambda I)\\). To do that, we need the eigenvalue \\(\lambda\\).

Step 1: To find the eigenvalue \\(\lambda\\), we must solve \\(\det(A - \lambda I) = 0\\). \
Step 1a: \\(A - \lambda I = \begin{bmatrix} 1 & 0 \\ 1 & 1\end{bmatrix} - \begin{bmatrix}\lambda & 0 \\ 0 & \lambda \end{bmatrix} = \begin{bmatrix} 1 - \lambda & 0 \\ 1 & 1 - \lambda\end{bmatrix}\\) \
Step 1b: \\(\det\left(\begin{bmatrix} 1 - \lambda & 0 \\ 1 & 1 - \lambda\end{bmatrix}\right) = (1 - \lambda)^2\\). \
Step 1c: \\((1 - \lambda)^2 = 0\\), so \\(\lambda = 1\\).

Step 2: now we can plug in \\(\lambda\\) and solve for \\(x\\). \
Step 2a: \\((A - \lambda I)x = (A - I)x = \left(\begin{bmatrix}1 & 0 \\ 1 & 1\end{bmatrix} - \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}\right)x = \begin{bmatrix}0 & 0 \\ 1 & 0\end{bmatrix}x = 0\\). \
Step 2b: we have already stipulated that \\(x\in\mathbb{R}^2\\), so \\(x^T = \begin{bmatrix} x_1 \\ x_2\end{bmatrix}\\). \
Step 2c: so \\(\begin{bmatrix}0 & 0 \\ 1 & 0\end{bmatrix}x = \begin{bmatrix}0 & 0 \\ 1 & 0\end{bmatrix}\begin{bmatrix} x_1 \\ x_2\end{bmatrix} = \begin{bmatrix} 0 \\ x_1\end{bmatrix} = 0\\).

Step 3: since step 2c shows that \\(x_1 = 0\\) but no equation involves \\(x_2\\), we know that \\(x_2\\) is free, so \\(x = \begin{bmatrix}0 & t\end{bmatrix}\\) where \\(t\in\mathbb{R}\\), so we can choose the basis vector \\(x = \begin{bmatrix}0 & 1\end{bmatrix}\\) as the eigenvector which spans our solution space. And we're done. 

b. Compute the eigenspace of: \\(B = \begin{bmatrix} -2 & 2 \\ 2 & 1\end{bmatrix}\\). 

Approach: this is just a slightly more complex version of 4.3a, so we will explain less.

Step 1a: \\(A - \lambda I = \begin{bmatrix}-2 - \lambda & 2 \\ 2 & 1 - \lambda\end{bmatrix}\\). \
Step 1b: \\(\det\left(\begin{bmatrix}-2 - \lambda & 2 \\ 2 & 1 - \lambda\end{bmatrix}\right) = (-2 - \lambda)(1 - \lambda) - 4\\). \
Step 1c: \\((-2 - \lambda)(1 - \lambda) - 4 = \lambda^2 +\lambda - 6 = 0\\). \
Step 1d: Factor \\((\lambda^2 +\lambda - 6 = (\lambda + 3)(\lambda - 2) = 0\\), so \\(\lambda_1 = -3\\) and \\(\lambda_2 = 2\\).

Step 2: we will start with finding the eigenvector \\(x\\) associated with \\(\lambda_1\\). \
Step 2a: \\(B - (-3)I = \begin{bmatrix} -2 & 2 \\ 2 & 1\end{bmatrix} - \begin{bmatrix}-3 & 0 \\ 0 & -3\end{bmatrix} = \begin{bmatrix}1 & 2 \\ 2 & 4\end{bmatrix}x = 0\\). \
Step 2b: \\(\begin{bmatrix}1 & 2 \\ 2 & 4\end{bmatrix}\begin{bmatrix}x_1 \\ x_2\end{bmatrix} = \begin{bmatrix}x_1 + 2x_2 \\ 2x_1 + 4x_2\end{bmatrix} = 0\\)
Step 2c: the first row is the only independent row, so \\(x_1 + 2x_2 = 0\\) and \\(x_1 = -2x_2\\). Pick a free parameter \\(t = x_2\\) giving \\(x = t\begin{bmatrix}-2 & 1\end{bmatrix}\\), which tells us that the eigenspace of \\(\lambda_1\\) is spanned by any scalar multiple of \\(\begin{bmatrix}-2 & 1\end{bmatrix}\\). And we're done. 

#### Exercise 4.4

Compute all eigenspaces of:
\\[
\begin{bmatrix}
0 & -1 & 1 & 1 \\
-1 & 1 & -2 & 3 \\
2 & -1 & 0 & 0 \\
1 & -1 & 1 & 0
\end{bmatrix}
\\]

Approach: We know that the eigenspace \\(E_\lambda\\) is the solution space for \\((A - \lambda I)x = 0\\). We need to find every non-trivial \\(x_i\in\mathbb{R}^4\\) s.t. \\(A - \lambda I = 0\\). So we need to find each eigenvalue \\(\lambda_i: i = 1\to 4\\). 

Step 1: we need to solve \\(\det(A - \lambda I) = 0\\).
Step 1a: therefore, we need to solve for \\(\lambda\\) in:

\\[\det\left(
\begin{bmatrix}
-\lambda & -1 & 1 & 1 \\
-1 & 1 - \lambda & -2 & 3 \\
2 & -1 & -\lambda & 0 \\
1 & -1 & 1 & -\lambda
\end{bmatrix}\right) = 0
\\]

Step 1b: to solve for \\(\lambda\\), we can do cofactor expansion or row reduction. We want to get to a triangular form, which will allow us to compute the determinant as \\(\det(A - \lambda I) = \prod(a_{ii} - \lambda)\\).

Step 1c: To do row reduction, we need to track determinant effects. Row swap scales the determinant by -1. Row scale by \\(k\\) scales the determinant the same way. Adding a multiple of a row does not change the determinant. 

Step 1c1: do \\(R_2\leftarrow\lambda R_2 - R_1\\). \\(\lambda R_2 = \begin{bmatrix}-\lambda & \lambda - \lambda^2 & -2\lambda & 3\lambda\end{bmatrix}\\). So:
\\[
\lambda R_2 - R_1 = \begin{bmatrix}-\lambda & \lambda - \lambda^2 & -2\lambda & 3\lambda\end{bmatrix} - \begin{bmatrix}0 - \lambda & -1 & 1 & 1\end{bmatrix} = \begin{bmatrix}-\lambda + \lambda & \lambda - \lambda^2 + 1 & -2\lambda - 1 & 3\lambda - 1\end{bmatrix}
\\]

Now we are at \\(\lambda\det\\) for:
\\[
\begin{bmatrix}
-\lambda & -1 & 1 & 1 \\
0 & \lambda - \lambda^2 + 1 & -2\lambda - 1 & 3\lambda - 1 \\
2 & -1 & -\lambda & 0 \\
1 & -1 & 1 & -\lambda
\end{bmatrix}
\\]

Step 1c2: do \\(R_3 \leftarrow\lambda R_3 + 2R_1\\), so:
\\[
\lambda R_3 + 2R_1 = 
\begin{bmatrix}2\lambda & -\lambda & - \lambda^2 & 0\end{bmatrix} + \begin{bmatrix}-2\lambda & -2 & 2 & 2\end{bmatrix} = \begin{bmatrix}0 & -\lambda - 2 & -\lambda^2 + 2 & 2\end{bmatrix}
\\]

This gives us \\(\lambda^2\det\\) for:
\\[
\begin{bmatrix}
-\lambda & -1 & 1 & 1 \\
0 & \lambda - \lambda^2 + 1 & -2\lambda - 1 & 3\lambda - 1 \\
0 & -\lambda - 2 & -\lambda^2 + 2 & 2 \\
1 & -1 & 1 & -\lambda
\end{bmatrix}
\\]
Step 1c3: do \\(R_4 \leftarrow \lambda R_4 + R_1\\):
\\[
\lambda R_4 + R_1 = 
\begin{bmatrix}\lambda & -\lambda & \lambda & -\lambda^2\end{bmatrix} + \begin{bmatrix}-\lambda & -1 & 1 & 1\end{bmatrix} = \begin{bmatrix}0 & -\lambda - 1 & \lambda + 1 & -\lambda^2 + 1\end{bmatrix}
\\]

This gives us \\(\lambda ^3\det\\) for:
\\[
\begin{bmatrix}
-\lambda & -1 & 1 & 1 \\
0 & \lambda - \lambda^2 + 1 & -2\lambda - 1 & 3\lambda - 1 \\
0 & -\lambda - 2 & -\lambda^2 + 2 & 2 \\
0 & -\lambda - 1 & \lambda + 1 & -\lambda^2 + 1
\end{bmatrix}
\\]

The key strategy (after which we're stopping because this is getting ridiculous): for a matrix \\(R\\), to eliminate entry \\(a\\) below pivot \\(p\\), do: \\(R_{\text{target}}\leftarrow p(\lambda)R_{\text{target}} - a\lambda R_{\text{pivot}}\\) (because \\(p(\lambda)a(\lambda) - a(\lambda)p(\lambda) = 0\\)).

Step 2 (in principle, because ridiculous): compute determinant using the product of the diagonal, which gives us the characteristic polynomial \\(p(\lambda)\\). 

Step 3: solve for \\(\lambda\\) in \\(p(\lambda) = 0\\). This gives us the eigenvalues. 

Step 4: for each eigenvalue \\(\lambda_i\\), solve for \\(v\\) in \\((A - \lambda_i I)v = 0\\) by row-reducing \\(A - \lambda_i I\\) to solve for free variables. The solution set is the eigenspace. 

#### Exercise 4.5

Check whether each of the following is diagonalizable and/or invertible:

\\[
A = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}
\\]

\\[
B = \begin{bmatrix}1 & 0 \\ 0 & 0\end{bmatrix}
\\]

\\[
C = \begin{bmatrix}1 & 1 \\ 0 & 1\end{bmatrix}
\\]

\\[
D = \begin{bmatrix}0 & 1 \\ 0 & 0\end{bmatrix}
\\]


Approach: Recall that for a matrix \\(A\\), \\(A^{-1}A = I\\). In the general case, we can solve for \\(A^{-1}\\) by writing down the augmented matrix \\(A|I\\) and using Gaussian elimination. Once it is in RREF, the right half will contain \\(A^{-1}\\). For a \\(2\times 2\\) matrix, invertibility is indicated by a non-zero determinant (MML 100). 

Step 1, invertibility: \\(A, C\\) are invertible (have determinant 1) and \\(B, D\\) are non-invertible. 

Step 2, diagonalizability: the characteristic polynomial for \\(A\in\mathbb{R}^{2\times 2}\\) is \\(p(\lambda) = \lambda^2 - \text{tr}(A)\lambda + \det(A)\\). For each matrix, we can solve for \\(\lambda\\). Here, we will use numpy.

```python
A = np.array([[1, 0], [0, 1]])
B = np.array([[1, 0], [0, 0]])
C = np.array([[1, 1], [0, 1]])
D = np.array([[0, 1], [0, 0]])

a = np.linalg.eig(A)
print(f"The eigenvalues and eigenvectors of A are {a.eigenvalues} and \n{a.eigenvectors}.")
print(f"Rank of A - L * I: {np.linalg.matrix_rank(A - a.eigenvalues[0] * np.eye(2))} and {np.linalg.matrix_rank(A - a.eigenvalues[1] * np.eye(2))}")
b = np.linalg.eig(B)
print(f"The eigenvalues and eigenvectors of B are {b.eigenvalues} and {b.eigenvectors}.")
c = np.linalg.eig(C)
print(f"The eigenvalues and eigenvectors of C are {c.eigenvalues} and {c.eigenvectors}.")
d = np.linalg.eig(D)
print(f"The eigenvalues and eigenvectors of D are {d.eigenvalues} and {d.eigenvectors}.")
```

Step 2a: there are three possible cases regarding the diagonalizability of \\(A\in\mathbb{R}^{2\times 2}\\): if there are two distinct eigenvalues, then the it is always diagonalizable. If the eigenvalues are repeated, then we must check whether the null space is 2-dimensional (i.e. \\(A=\lambda I\\) (diagonalizable), or 1-dimensional (not diagonalizable). We can figure this out by finding out the rank of \\(A - \lambda I\\). Symmetric matrices are always diagonalizable.

Step 2b: Given the last point (symmetricity), \\(A, B\\) are diagonalizable. 

Step 2c: let's check rank of \\(A - \lambda I\\) for \\(C, D\\).

```python
c_diag = np.linalg.matrix_rank(C - c.eigenvalues[0] * np.eye(2))
d_diag = np.linalg.matrix_rank(D - d.eigenvalues[0] * np.eye(2))

print(f"C is diagonalizable {c_diag}: {c_diag == 0}")
print(f"D is diagonalizable: {d_diag == 0}")
```

An alternative approach is \\(\Delta = (\text{tr}(A))^2 - 4\det(A)\\). Cases:

- \\(\Delta > 0\\): diagonalizable
- \\(\Delta = 0\\): repeated, must check eigenspace (i.e. rank of \\(A - \lambda I\\).
- \\(\Delta < 0\\): not diagonalizable over \\(\mathbb{R}\\).

Step 1d: normalize by \\(Av\\): \\((AA^T) = \lambda \\). \
